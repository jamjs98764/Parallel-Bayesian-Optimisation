# L2 Error against Number of Iterations

```python
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
def error_vs_iterations(func = "egg", metrics = "L2", batch = False, batch_size = 2, heuristic = "kb", aggr_seed = "mean"):
    if batch == False:
        filename = func + "FITBOMM_results_" + metrics + ".npy"
        results = np.load(filename)
    else:
        filename = func + "FITBOMM_results_" + metrics + "," + str(batch_size) + "_batch_size," + heuristic + "_heuristic.npy"
        results = np.load(filename)
        results = np.repeat(results, repeats = batch_size * np.ones(results.shape[1], dtype = int), axis = 1)
    
    df = pd.DataFrame(results)
    df = df.stack()
    df = df.to_frame()
    df.index.names = (['seed', 'iters'])
    df.columns = ['values']
    df.reset_index(inplace = True)
    return df
    
    #return results, seed_size, num_iterations
```

```python
results1 = error_vs_iterations(batch = False)    
results2 = error_vs_iterations(batch = True, heuristic = "random")

fig = sns.lineplot(x = 'iters', y = 'values', data = results1, err_style = "band", legend = 'Sequential')
fig = sns.lineplot(x = 'iters', y = 'values', data = results2, err_style = "band", legend = 'random batch')  
```
