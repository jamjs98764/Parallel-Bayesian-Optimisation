## Filter features before learning algorithm

To filter features before applying the learning algorithm, you can measure some sort of correlation between each feature and the output. Pearson correlation works well when both input and output are numeric. 

I did not find any practical way to perform feature selection with the categorical variables. 


```python
categorical_feats = ["APEX_TYPE", "GENDER", "MARITALSTATUS", "KIDSINHH", "EDUCATION", "INCOME", "NIELSEN_COUNTY_SIZE", "BDSSEGMENT"]
numerical_feats = ["AGE", "LOYALTY", "PRICESENSITICITY", "BRANDTRIAL", "BRANDSHOPPING", "BRANDOPENNESS"]
```

### Pearson correlation 
Compute correlation with respect to the `EMAILSOPENED_RATIO` column, using the numerical features in the `APEX_demo` dataset.

In any case, none of the several features has a big pearson correlation. 

```python
cor = APEX_demo[numerical_feats].corrwith(APEX_engagement["EMAILSOPENED_RATIO"])
```

### Analysis: Emails Opened

Trying K-prototype clustering

```python
ordinal_df = pd.read_csv("joined_data_with_ordinal_categories.csv")
```

```python
ordinal_df.set_index("CONSUMERID")
print(ordinal_df.columns)
```

```python
y_features = ['EMAILSSENT', 'EMAILSOPENED', 'EMAILSCLICKED',
       'opened_perc', 'clicked_perc', 'has_opened_email', 'has_clicked_email']
x_features = ['APEX_TYPE', 'GENDER', 'AGE', 'MARITALSTATUS',
       'KIDSINHH', 'BDSSEGMENT', 'LOYALTY', 'PRICESENSITICITY', 'BRANDTRIAL',
       'BRANDSHOPPING', 'BRANDOPENNESS', 'INCOME_cat', 'EDUCATION_cat',
       'NIELSEN_COUNTY_SIZE_cat']
```

```python
from kmodes.kprototypes import KPrototypes

# Printing index of features
for i in range(len(x_features)):
    print(x_features[i] + ": " + str(i))
```

```python
x_categories = [0, 1, 3, 4, 5]
# x_categories = [0, 1, 3, 4]
x_ordinal = [11, 12, 13]
```

```python
df_y = ordinal_df[y_features]
df_x = ordinal_df[x_features]

df_x.dtypes
```

```python
# df_x = df_x.head(100) # Test with smaller df
```

```python
np.random.seed(36)
kproto = KPrototypes(n_clusters = 5, init = "Huang", verbose = 1) # Change verbose for displaying calculations
clusters = kproto.fit_predict(df_x.values, categorical = x_categories)
```

```python
clusters
```

```python
print("Numerical feature centroids \n")
print(kproto.cluster_centroids_[0])
print("\n Categorical feature centroids \n")
print(kproto.cluster_centroids_[1])

print("\n Instance labels \n")
print(kproto.labels_)

# Print training statistics
print(kproto.cost_)
print(kproto.n_iter_)
```

```python
import pickle
file_name = "kproto_5_huang.pickle"
pickle_dict = {
        "model": kproto, 
        "clusters": clusters,
        }

with open(file_name, 'wb') as f:
    pickle.dump(pickle_dict, f)   
```

Plotting cost vs. number of clusters. Note, we use random sampling to reduce df size for easier computation first

```python
df_small = df_x.sample(frac = 0.2, random_state = 22)
```

```python
k_range = np.arange(1,10)
cost_record = []
for k in k_range:
    kproto = KPrototypes(n_clusters = k, init = "Huang", verbose = 0)
    clusters = kproto.fit_predict(df_small.values, categorical = x_categories)
    cost_record.append(kproto.cost_)
```

```python
plt.plot(k_range, cost_record)
```
